```python
#i creat all of the data processing as a def cause i think  same way to handle it
def split_komoditi(table_name,x,y,df):
    df_BM = df.iloc[1:,[0]+list(range(x, y+1))]
    df_BM.iloc[0,0] = "Province"
    for i in df_BM.columns:
        if i != "Provinsi":
            df_BM[i]=df_BM[i].fillna(0).astype(int)
    new_header = df_BM.iloc[0]
    df = df_BM.iloc[2:]
    df.columns = new_header

    df = df.reset_index(drop=True)
    df.to_sql(table_name, engine, index=False, if_exists='replace')
    print("success..")
    
```

```python
#usage of
x = 21
y =25
table_name = "tomat"

split_komoditi(table_name,x,y,df)
```
